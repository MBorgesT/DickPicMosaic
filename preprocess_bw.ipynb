```python
import pandas as pd
import numpy as np
import cv2
import os
from alive_progress import alive_bar
```

```python
rows = []
class_size = 5
n_classes = int(255/class_size)
for i in range(n_classes):
    rows.append([i * class_size, 0, 'none'])

df = pd.DataFrame(rows, columns=['bw', 'n_files', 'files'])
df.head()
```

```python
files = os.listdir('resized_pics/')
with alive_bar(len(files)) as bar:
    for f in files:
        img = cv2.imread(f'resized_pics/{f}', cv2.IMREAD_GRAYSCALE)
        avg_color_per_row = np.average(img, axis=0)
        avg_color = np.average(avg_color_per_row, axis=0)

        color_class = avg_color - avg_color % class_size
        index = (color_class / class_size)
        index = int(index)

        files = df.loc[index, 'files']
        files = files.split(';') if files != 'none' else []
        files.append(f)
        df.at[index, 'files'] = ';'.join(files)
        df.at[index, 'n_files'] = len(files)

        bar()
```
